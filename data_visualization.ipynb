```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/bengaluru_house_prices.csv')
```

```python
#print first five rows
df.head()
```

```python
#check the datatypes
df.dtypes
```

```python
#statistical summary
df.describe()
```

```python
#check location column exist
df = df.reset_index()
df = df.set_index('location')
```

Line Plot

```python
# Assuming the 'location' column is the third column (index 2) in your DataFrame

# Increase the figure size and adjust the width
plt.figure(figsize=(12, 6))

df.plot(x=df.columns[2], y='price', kind='line')
plt.title('Price of the House by Location')
plt.ylabel('Price')
plt.xlabel('Location')

plt.show()

# Save the plot as an image
plt.savefig('plot.png')
```

```python
df.sort_values(['price'], ascending = False, axis = 0, inplace = True)
print(df['price'])
```

Area Plot

```python
df.plot(x=df.columns[2], y='price', kind='area')
plt.title('Price of the house based on Location')
plt.xlabel('Location')
plt.ylabel('Price')
plt.show()
```

Histogram

```python
df.plot(x=df.columns[2], y='price', kind='hist')
plt.title('Price of the house based on Location')
plt.xlabel('Location')
plt.ylabel('Price')
plt.show()
```

Bar Chart

```python
df.plot(x='size', y='price', kind='bar')
plt.title('Price of the house based on Size')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```

Box Plot

```python
df.plot(x=df.columns[2], y='price', kind='box')
plt.title('Price of the house based on Size')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```
